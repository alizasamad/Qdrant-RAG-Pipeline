## Synthetic Dataset Evaluation by AWS

Run this script for all three methods to generate comparable test results. This will generate groundedness & relevancy scores that you can use to statistically evaluate model performance. Ensure to keep sample size constant among all 3 datasets. This is important during step 2: evaluate_models.py.

```python
#!pip install -r requirements.txt
```

```python
# adjustable parameters
method = 'aws'  # method used to synthesize doc ['aws', 'ragas', 'naive']
doc = 'swa' # name/abbrv of doc used
target = 'google_NQ_simplified.csv'  # dataset being evaluated
n = 50  # how many q/a pairs to sample & evaluate from target
```

```python
from AWS_data_creation import evaluate_dataset
import pandas as pd
import random
import os

# read in the synthetic RAG dataset
os.chdir('../')
data = pd.read_csv(f"datasets/{target}")
os.chdir('test_data_synth/')

# sample 'n' q/a pairs to evaluate
random.seed(123)
indices = random.sample(range(len(data)), n)
data_sample = data.iloc[indices]
data_sample
```

### Running Dataset Evaluation

This process can take over 20 minutes depending on how large your sample size is.

```python
dataset_evaluated = evaluate_dataset(data_sample, method)
dataset_evaluated.head()
```

### Save Your Results

You will find your results in the 'results' folder of the 'test_data_synth' working directory.

```python
dataset_evaluated.to_csv(f"results/test_results_{method}_synth_{str(n)}_{doc}.csv", index=False)
```
